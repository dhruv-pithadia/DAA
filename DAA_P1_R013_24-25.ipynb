```python
import random
import time
import matplotlib.pyplot as plt

def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Generate arrays of 10, 100, 1000, 2000, 5000, 10000, and 50000 random numbers
arr10 = [random.randint(0, 10) for _ in range(10)]
arr100 = [random.randint(0, 100) for _ in range(100)]
arr1000 = [random.randint(0, 1000) for _ in range(1000)]
arr2000 = [random.randint(0, 2000) for _ in range(2000)]
arr5000 = [random.randint(0, 5000) for _ in range(5000)]
arr10000 = [random.randint(0, 10000) for _ in range(10000)]
arr50000 = [random.randint(0, 50000) for _ in range(50000)]
arr100000 = [random.randint(0, 50000) for _ in range(100000)]

# Measure the time taken to sort each array
start_time = time.time()
insertionSort(arr10)
itime10 = time.time() - start_time

start_time = time.time()
insertionSort(arr100)
itime100 = time.time() - start_time

start_time = time.time()
insertionSort(arr1000)
itime1000 = time.time() - start_time

start_time = time.time()
insertionSort(arr2000)
itime2000 = time.time() - start_time

start_time = time.time()
insertionSort(arr5000)
itime5000 = time.time() - start_time

start_time = time.time()
insertionSort(arr10000)
itime10000 = time.time() - start_time

start_time = time.time()
insertionSort(arr50000)
itime50000 = time.time() - start_time

start_time = time.time()
insertionSort(arr100000)
itime100000 = time.time() - start_time

# Print the times
print("Time taken to sort 10 elements: ", itime10)
print("Time taken to sort 100 elements: ", itime100)
print("Time taken to sort 1000 elements: ", itime1000)
print("Time taken to sort 2000 elements: ", itime2000)
print("Time taken to sort 5000 elements: ", itime5000)
print("Time taken to sort 10000 elements: ", itime10000)
print("Time taken to sort 50000 elements: ", itime50000)
print("Time taken to sort 50000 elements: ", itime100000)

# Plotting the results with larger figure size
plt.figure(figsize=(10, 6))

sizes = [10, 100, 1000, 2000, 5000, 10000, 50000, 100000]
itimes = [itime10, itime100, itime1000, itime2000, itime5000, itime10000, itime50000, itime100000]

plt.plot(sizes, itimes, marker='o')
plt.xlabel('Array Size')
plt.ylabel('Time Taken (seconds)')
plt.title('Insertion Sort Time Complexity')

# Adding text annotations
for i, txt in enumerate(itimes):
    plt.annotate(f"{txt:.6f}", (sizes[i], itimes[i]), textcoords="offset points", xytext=(0,5), ha='center')

plt.grid(True)
plt.show()
```

```python
def insertionSortDescending(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key > arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Example usage
arr = [12, 11, 13, 5, 6]
insertionSortDescending(arr)
print("Sorted array in descending order: ", arr)

insertionSort(arr)
print(arr)
```

```python
#Bubble sort
import random
import time
import matplotlib.pyplot as plt

def bubbleSort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Generate arrays of 10, 100, 1000, 2000, 5000, 10000, and 50000 random numbers
arr10 = [random.randint(0, 10) for _ in range(10)]
arr100 = [random.randint(0, 100) for _ in range(100)]
arr1000 = [random.randint(0, 1000) for _ in range(1000)]
arr2000 = [random.randint(0, 2000) for _ in range(2000)]
arr5000 = [random.randint(0, 5000) for _ in range(5000)]
arr10000 = [random.randint(0, 10000) for _ in range(10000)]
arr50000 = [random.randint(0, 50000) for _ in range(50000)]
arr100000 = [random.randint(0, 50000) for _ in range(100000)]

# Measure the time taken to sort each array using bubble sort
start_time = time.time()
bubbleSort(arr10)
time10 = time.time() - start_time

start_time = time.time()
bubbleSort(arr100)
time100 = time.time() - start_time

start_time = time.time()
bubbleSort(arr1000)
time1000 = time.time() - start_time

start_time = time.time()
bubbleSort(arr2000)
time2000 = time.time() - start_time

start_time = time.time()
bubbleSort(arr5000)
time5000 = time.time() - start_time

start_time = time.time()
bubbleSort(arr10000)
time10000 = time.time() - start_time

start_time = time.time()
bubbleSort(arr50000)
time50000 = time.time() - start_time

start_time = time.time()
bubbleSort(arr100000)
time100000 = time.time() - start_time

# Print the times
print("Time taken to sort 10 elements: ", time10)
print("Time taken to sort 100 elements: ", time100)
print("Time taken to sort 1000 elements: ", time1000)
print("Time taken to sort 2000 elements: ", time2000)
print("Time taken to sort 5000 elements: ", time5000)
print("Time taken to sort 10000 elements: ", time10000)
print("Time taken to sort 50000 elements: ", time50000)
print("Time taken to sort 100000 elements: ", time100000)

# Plotting the results with larger figure size
plt.figure(figsize=(10, 6))

sizes = [10, 100, 1000, 2000, 5000, 10000, 50000, 100000]
times = [time10, time100, time1000, time2000, time5000, time10000, time50000, time100000]

plt.plot(sizes, times, marker='o')
plt.xlabel('Array Size')
plt.ylabel('Time Taken (seconds)')
plt.title('Bubble Sort Time Complexity')

# Adding text annotations
for i, txt in enumerate(times):
    plt.annotate(f"{txt:.6f}", (sizes[i], times[i]), textcoords="offset points", xytext=(0,5), ha='center')

plt.grid(True)
plt.show()
```
