```python
import time
import matplotlib.pyplot as plt

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Test cases
test_cases = [
    [], 
    [1], 
    [3, 6, 8, 10, 1, 2, 1],  
    [5, 5, 5, 5],  
    [1, 2, 3, 4, 5],  
    [9, 8, 7, 6, 5],  
]


times = []
sizes = []

for test in test_cases:
    start_time = time.time()
    quick_sort(test)
    elapsed_time = time.time() - start_time
    times.append(elapsed_time)
    sizes.append(len(test))

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(sizes, times, marker='o', color='b', label="Quick Sort Time Complexity")
plt.title("Time Complexity of Quick Sort for Different Test Cases")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.legend()
plt.show()

times, sizes
```

```python
import random
import time
import matplotlib.pyplot as plt

# Randomized Quick Sort Implementation
def partition(arr, lo, hi):
    pivot = arr[hi]
    i = lo
    for j in range(lo, hi):
        if arr[j] <= pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[hi] = arr[hi], arr[i]
    return i

def partition_r(arr, lo, hi):
    r = random.randint(lo, hi)
    arr[r], arr[hi] = arr[hi], arr[r]
    return partition(arr, lo, hi)

def quick_sort_randomized(arr, lo, hi):
    if lo < hi:
        p = partition_r(arr, lo, hi)
        quick_sort_randomized(arr, lo, p - 1)
        quick_sort_randomized(arr, p + 1, hi)

# Normal Quick Sort Implementation
def quick_sort_normal(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort_normal(left) + middle + quick_sort_normal(right)

# Function to measure time for both algorithms
def measure_time(test_cases):
    random_times = []
    normal_times = []

    # Measure time for Randomized Quick Sort
    for test in test_cases:
        arr_copy = test.copy()
        start_time = time.time()
        quick_sort_randomized(arr_copy, 0, len(arr_copy) - 1)
        elapsed_time = time.time() - start_time
        random_times.append(elapsed_time)

    # Measure time for Normal Quick Sort
    for test in test_cases:
        arr_copy = test.copy()
        start_time = time.time()
        quick_sort_normal(arr_copy)
        elapsed_time = time.time() - start_time
        normal_times.append(elapsed_time)

    return random_times, normal_times

# Generate large test cases
sizes = [100, 1000, 5000, 10000, 50000, 100000]
test_cases = [[random.randint(0, 100000) for _ in range(size)] for size in sizes]

# Measure the time complexities for both algorithms
random_times, normal_times = measure_time(test_cases)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, random_times, marker='o', label="Randomized Quick Sort", color='r')
plt.plot(sizes, normal_times, marker='x', label="Normal Quick Sort", color='b')
plt.title("Time Complexity Comparison: Randomized Quick Sort vs Normal Quick Sort")
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid(True)
plt.show()

# Print the time complexities for both algorithms
for i, size in enumerate(sizes):
    print(f"Input Size: {size}, Randomized Quick Sort Time: {random_times[i]:.6f}s, Normal Quick Sort Time: {normal_times[i]:.6f}s")
```
